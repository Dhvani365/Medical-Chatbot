```python
from langchain.documents_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

## Creating documents from book

```python
def load_pdf(pdf_file):
    loader = DirectoryLoader(
        pdf_file,
        glob="**/*.pdf",
        loader_cls=PyPDFLoader
    )

    documents = loader.data()
    return documents
```

```python
extracted_data = load_pdf("data")
```

```python
from typing import List
from langchain.schema import Document

def filter_to_minimal_doc(docs: List[Document]) -> List[Document]:
    minimal_docs = []
    for doc in docs:
        minimal_doc = Document(
            page_content=doc.page_content,
            metadata={"source": doc.metadata.get("source", "")}
        )
        minimal_docs.append(minimal_doc)
    return minimal_docs
```

```python
minimal_docs = filter_to_minimal_doc(extracted_data)
```

## Creating Chunks from docs

```python
def text_split(minimal_docs):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=20,
    )
    text_chunks = text_splitter.split_documents(minimal_docs)
    return text_chunks
```

```python
docs_chunk = text_split(minimal_docs)
print(f"number of chunks: {len(docs_chunk)}")
```
